```r
x = runif(min=0,max=10,n=10)
y = 3+2*x+rnorm(n=10,mean=0,sd=5)
x=round(x,2)
y=round(y,2)
plot(x~y)
```

```r
model = lm(x~y)
```

```r
plot(x~y,col='cyan4')
abline(model,col='red')
```

```r
intercept = coef(model)[1]
slope = coef(model)[2]
y.hat = intercept + slope*x
round(y.hat,2)
r = y-y.hat
round(r,2)
```
