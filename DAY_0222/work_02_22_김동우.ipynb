```r
remove_outlier = function(data,method,lim = 1.5,threshold=3){
    if(method =='iqr'|method=='IQR'){
        q = quantile(data,probs = c(0.25,0.75),na.rm=TRUE)
        iqr = diff(q)
        lower = q[1] - lim*iqr
        upper = q[2] + lim*iqr
        output=data[!(data<lower | data>upper)]
        
        return(output)
    }
    else if(method =='z'|method=='Z'|method=='z-score'){
        z_score = scale(data)
        output =  data[!abs(z_score)>threshold]

        return(output)

    }
    else{'wrong input'}

}
```
