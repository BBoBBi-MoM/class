```r
library(dplyr)
```

```r
get_frequency_table = function(num){
    frequency_table = table(sample(1:6,num,replace=T))
    return(frequency_table)
}
```

```r
iter_12 = get_frequency_table(12)
iter_1200 = get_frequency_table(1200)
iter_12000 = get_frequency_table(12000)

table = bind_rows(iter_12,iter_1200,iter_12000)
table = table[,order(names(table))]
table
par(mfrow=c(3,1))
barplot(iter_12,main='12')
barplot(iter_1200,main='1200')
barplot(iter_12000,main='12000')
```

```r
get_binom_hist = function(prob,trial=10){
    x = 1:10
    y = dbinom(x,trial,prob)

    plot(x,y,type='l')
    cat('P:',prob)
    axis(side=1,at=1:10)
}
get_binom_hist(0.5)
get_binom_hist(0.1)
get_binom_hist(0.9)
```
