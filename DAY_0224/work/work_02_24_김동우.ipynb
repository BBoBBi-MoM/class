```r
source('../../utils/util.r')
height_weight = read.csv('height_weight.csv')
```

```r
display_info(height_weight)
plot(height_weight$weight~height_weight$height)
```

```r
bmi = (height_weight$weight)/(height_weight$height/100)^2
height_weight_bmi = cbind(height_weight,bmi)
height_weight = height_weight_bmi[15<height_weight_bmi$bmi&height_weight_bmi$bmi<35,]
plot(height_weight$weight,height_weight$height,col='cyan4')
```

```r
simple_linear_model = lm(height_weight$height~height_weight$weight)
plot(height_weight$weight,height_weight$height,col='cyan4')
abline(simple_linear_model,col='red')
```
