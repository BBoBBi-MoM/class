```r
#가설
# H0: 아프리카 대륙과 다른 지역 간의 에이즈 발병률은 차이 없다.
# H1: 아프리카 대륙의 에이즈 발병률은 다른 지역에 비해 높다.
```

```r
# 세계 에이즈 통계,세계 인구통계
hiv_data = read.csv('./archive/HIV.csv')
pop_data = read.csv('./archive/population_by_country_2020.csv')
hiv_data
```

```r
sum(is.na(hiv_data))
```

```r
#인구데이터 정제
pop_data = pop_data[,1:2]
colnames(pop_data) = c('country','pop')
```

```r
#에이즈데이터 정제
hiv_country = hiv_data[1]
hiv_median = hiv_data[5]
hiv_continent = hiv_data[ncol(hiv_data)]
hiv_data = cbind(hiv_country,hiv_median,hiv_continent)
colnames(hiv_data) = c('country','median','continent')
```

```r
hiv_df
table(hiv_df$continent)
```

```r
#두개의 데이터 프레임 합친 후 전처리
hiv_df = merge(pop_data,hiv_data,by='country')
hiv_df = na.omit(hiv_df)
hiv_rate = hiv_df[3]/hiv_df[2]*100
colnames(hiv_rate) = 'hiv_rate'

hiv_df = cbind(hiv_df,hiv_rate)
sorted_by_rate = hiv_df[order(hiv_df$hiv_rate,decreasing=TRUE),]
top_50 = sorted_by_rate[1:50,]

table_50 = table(top_50$continent)
pie(table_50,col = c('red','orange','green','cyan','skyblue','purple'),main='에이즈 발병률 상위 50개 국가의 대륙 분포')
table_50
top_30 = sorted_by_rate[1:30,]

table_30 = table(top_30$continent)
pie(table_30,col = c('red','orange','green','cyan','skyblue','purple'),main='에이즈 발병률 상위 30개 국가의 대륙 분포')

top_10 = sorted_by_rate[1:10,]

table_10 = table(top_10$continent)
pie(table_10,col = c('red','orange','green','cyan','skyblue','purple'),main='에이즈 발병률 상위 10개 국가의 대륙 분포')
```

```r
#아프리카 데이터프레임 생성
africa_hiv_df = hiv_df[hiv_df$continent == 'Africa',1:5]
africa_hiv_df
```

```r
#아프리카 에이즈 발병률의 통계
africa_rate_mean = mean(africa_hiv_df$hiv_rate)
cat('아프리카 에이즈 발병률 평균:',africa_rate_mean,'\n')
africa_rate_var = var(africa_hiv_df$hiv_rate)
cat('아프리카 에이즈 발병률 분산:',africa_rate_var)
```

```r
#비 아프리카 데이터프레임 생성
non_africa_hiv_df = hiv_df[hiv_df$continent != 'Africa',]
non_africa_hiv_df
```

```r
africa_rate = africa_hiv_df$hiv_rate
africa_rate
```

```r
non_africa_rate = none_africa_hiv_df$hiv_rate
non_africa_rate
```

```r
t_val = 4.1191
```

```r
x = seq(-3,4.5,length = 200)
plot(x,dnorm(x),type='l')
abline(v=t_val,col='red')
abline(v = qt(0.95,41),col='blue')
```

```r
t.test(africa_rate,non_africa_rate,alternative='greater',var=FALSE)
```

### 검정결과
