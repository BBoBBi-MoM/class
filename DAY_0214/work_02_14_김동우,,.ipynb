##  R에서 데이터 자료형 분류

1. 스칼라(Scala)  
구성인자가 하나인 벡터  
  
2. 벡터(Vector)  
동일한 유형의 데이터가 1차원으로 구성되어 있는 데이터 구조  
c()
  
3. 팩터, 요인(Factor)  
범주형데이터(자료)를 표현하기 위한 데이터 타입  
factor()
  
4. 행렬(Matrix)  
동일한 유형의 2차원 데이터 구조 (예) : (행(Row), 열(Column))  
matrix()  
  
5. 배열(Array)  
배열은 동일한 유형의 다차원 데이터 구조  
array()
  
6. 리스트(List)  
리스트는 서로 다른 구조의 데이터를 모두 묶은 객체  
list()
7. 데이터프레임(Data.frame)  
데이터 유형에 상관없이 2차원 형태의 데이터 구조  
data.frame()

## 데이터 불러오기, 데이터 전처리

```r
library(readxl)
raw_data = read.csv("2020_raw_data.csv",fileEncoding = "euc-kr")
colnames(raw_data)<-c('지역','가족원합계','세대구성','가구구성','교육정도')
meta_data = read_excel("meta_data.xlsx")
```

메타 데이터 정제

```r
meta_df = data.frame(meta_data) # 리스트를 데이터 프레임으로
names(meta_df)=meta_df[2,] # 컬럼 네임 재설정
meta_df = meta_df[c(-1,-2),c(-8,-7,-6,-5)] # 불필요한 정보 제거
meta_df
```

---

## 3-1 지역

지역정보 정제

```r
location_df = subset(meta_df,meta_df$연번 == 2) # 지역이름 데이터 프레임
location_label = location_df[,4]
location_code = as.integer(location_df[,3])
location_data = factor(raw_data$지역,levels = location_code, labels = location_label)
```

세대정보 정제

```r
generation_df = subset(meta_df,meta_df$연번 == 7) # 세대구성
generation_label = generation_df[,4]
generation_code = as.integer(generation_df[,3])
generation_data = factor(raw_data$세대구성,levels = generation_code, labels = generation_label)
```

가구구성정보 정제

```r
household_df = subset(meta_df,meta_df$연번 == 8) # 가구구성
household_label = household_df[,4]
household_code = as.integer(household_df[,3])
household_data = factor(raw_data$가구구성,levels = household_code, labels = household_label)
```

지역&세대,지역&가구 테이블 생성


```r
loc_gen_table = table(location_data,generation_data)
loc_house_table = table(location_data,household_data)
```

지역&세대 시각화 (파이차트)

```r
for(i in seq(1,16,2)){
    par(mfrow=c(1,2))
    pie(loc_gen_table[i,],main=location_label[i],radius=1)
    pie(loc_gen_table[i+1,],main=location_label[i+1],radius=1)
}
```

지역&가구 시각화 (파이차트)

```r
for(i in seq(1,16,2)){
    par(mfrow=c(1,2))
    pie(loc_house_table[i,],main=location_label[i],radius=1)
    pie(loc_house_table[i+1,],main=location_label[i+1],radius=1)
}
```

```r
cat('가장 많은 세대구성:',names(which.max(table(generation_data))),'\n')
cat('가장 많은 가구구성:',names(which.max(table(household_data))),'\n\n')
cat('가장 적은 세대구성:',names(which.min(table(generation_data))),'\n')
cat('가장 적은 가구구성:',names(which.min(table(household_data))),'\n\n')
cat('평균 세대구성:',round(mean(raw_data$세대구성),2),'\n')
cat('평균 가구구성:',round(mean(raw_data$가구구성),2),'\n')
```

## 3-2 가족원 합계

```r
num_of_household = table(raw_data$가족원합계)
cat("가장 많은 가족원 수",labels(which.max(num_of_household)),'\n')
cat("가장 적은 가족원 수",labels(which.min(num_of_household)),'\n')
cat("평균 가족원 수",round(mean(raw_data$가족원합계),2),'\n')
```
