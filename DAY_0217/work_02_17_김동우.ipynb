```r
pop_mean_inference = function(sample_mean,sample_sd,sample_size,pop_sd,unit,alpha=0.05){
    lower = sample_mean - qnorm(1-(alpha/2))*(pop_sd/sqrt(sample_size))
    upper = sample_mean + qnorm(1-(alpha/2))*(pop_sd/sqrt(sample_size))
    x = seq(sample_mean-3*sample_sd,sample_mean+3*sample_sd,0.01)
    y = dnorm(x,mean=sample_mean,sd=sample_sd)
    plot(x,y,type='l')
    title(xlab=unit,main='모평균 추정')
    x_lim = seq(lower,upper,0.01)
    y_lim = dnorm(x_lim,mean=sample_mean,sd=sample_sd)

    x_lim = c(lower,x_lim,upper)
    y_lim = c(0,y_lim,0)

    polygon(x_lim,y_lim,col='red')
    cat('모평균은 신뢰수준',(1-alpha)*100,'%에서 ',lower,unit,'~',upper,unit,'구간에 있을거라 추정된다')
}
```

```r
mean_of_A = 12
mean_of_B = 8
sd_of_A = 2
sd_of_B = 3

sample_size_A = 8
sample_size_B = 10

t_value = (mean_of_A - mean_of_B) / sqrt((sd_of_A^2/sample_size_A) + (sd_of_B^2/sample_size_B))

p_value = 2 * pt(abs(t_value), df = sample_size_A + sample_size_B - 2, lower.tail = FALSE)

cat("t-value: ", t_value, "\n")
cat("p-value: ", p_value, "\n")
```

```r
pop_mean_inference(product_sample_mean,product_sample_sd,product_sample_size,product_sd,unit='년')
```
