```python
boston_df = boston_df.sample(frac=1)
train_data , test_data = boston_df.iloc[:int(len(boston_df)*0.8),:], boston_df.iloc[int(len(boston_df)*0.8):,:]
train_X,train_y = train_data.iloc[:,:-1].values , train_data.iloc[:,-1].values 
test_X,test_y = test_data.iloc[:,:-1].values , test_data.iloc[:,-1].values 
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(train_X,train_y)
prediction = model.predict(test_X)
```

```python
model.coef_, model.intercept_
```

```python
np.sqrt(((test_y - prediction)**2).mean())
```
