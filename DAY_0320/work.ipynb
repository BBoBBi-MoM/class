```python
from sklearn.datasets import load_digits
from sklearn.preprocessing import StandardScaler
from sklearn.decomposition import PCA

import matplotlib.pyplot as plt
import pandas as pd
```

```python
digits = load_digits()
digits_label = digits.target
```

```python
digits_data = pd.DataFrame(digits.data)
digits_label = pd.DataFrame(digits_label)
digits_label.columns = ['label']
digits_df = pd.concat([digits_data,digits_label],axis=1)
```

```python
digits_df = digits_df[(digits_df['label']==1) |(digits_df['label']==3 )| (digits_df['label']==9)]
digits_df
```

```python
features = StandardScaler().fit_transform(digits_df.iloc[:,:-1])
```

```python
features
```

```python
pca = PCA(n_components=0.99, whiten=True)
features_pca = pca.fit_transform(features) # PCA를 수행

print("원본 특성 개수:", features.shape[1]) # 결과를 확인
print("줄어든 특성 개수:", features_pca.shape[1])
```
